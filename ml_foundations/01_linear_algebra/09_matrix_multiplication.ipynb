# Matrix Multiplication



Matrix B's row number must be equal to Matrix A's column number

## Matrix by Vector Multiplication

```python
import numpy as np
A = np.array([[3,4],[5,6],[7,8]])
A
```

```python
b = np.array([1,2])
b
```

```python
np.dot(A,b)
```

```python
import torch 
A_pt = torch.tensor([[3,4],[5,6],[7,8]])
A_pt
```

```python
b_pt = torch.tensor([1,2])
b_pt
```

```python
torch.matmul(A_pt, b_pt) #automatically infers dimensions in order to perform dot product, matvec or matrix multiplication
```
