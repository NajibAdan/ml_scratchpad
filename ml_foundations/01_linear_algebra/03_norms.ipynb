# L<sup>2</sup> Norm


- Measures simple distance from origin  

```python
import numpy as np
x = np.array([25,2,5])
x
```

L<sup>2</sup> Norm is calculated as squaring all the elements in the vector then summing it and square rooting it

```python
sum = 0
for i in x:
    sum += i ** 2
l2_norm = sum **(1/2)
l2_norm # l2_norm is also the euclidean distance!!
```

Or we can use the inbuilt function in numpy

```python
np.linalg.norm(x)
```

# Unit Vectors

# L<sup>1</sup> Norm


It varies linearly at all locations whether near or far from origin. Used whenever difference between zero and non-zero is key

```python
sum = 0
for i in x:
    sum += np.abs(i)
sum
```

# Squared L<sup>2</sup> Norm


- Computationally cheap bcos: 
    - squared L<sup>2</sup> norm is X<sup>T</sup>X
    - Derivative of element x requires that element alone where as L<sup>2</sup> norm requires x vector

- Con: grows slowly nearly the origin can't be used to distinguish between zero and near-zero 

```python
sum = 0
for i in x:
    sum += np.square(i) 

sum
```

```python
np.dot(x,x)
```

# Max Norm (or L<sup>∞</sup>)



- Returns the absolute value of the largest magnitude element

```python
np.max(np.array([abs(i) for i in x]))
```
