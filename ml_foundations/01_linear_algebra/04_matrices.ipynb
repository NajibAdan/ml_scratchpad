# Matrices

## Matrices (Rank 2 Tensors) in Numpy

```python
import numpy as np
X = np.array([[25,2],[5,26],[3,7]])
X
```

```python
X.shape
```

```python
X.size
```

```python
print(X[:,0]) # Select left column
print(X[1,:]) # Select middle row
print(X[0:2, 0:2]) # Slicing by index
```

## Matrice in PyTorch

```python
import torch 

X_pt = torch.tensor([[25,2],[5,26],[3, 7]])
X_pt.shape
```

```python
X_pt[1,:]
```
