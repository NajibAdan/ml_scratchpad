# Frobenius Norm

It measures the size of matrix in terms of Euclidean distance
It's the sum of the magnitude of all the vectors in X.



```python
import numpy as np 

X = np.array([[1,2],[3,4]])

X
```

```python
sum = 0
for i in X:
    for j in i:
        sum += j ** 2
frob_norm = sum ** (1/2)
frob_norm
```

```python
np.linalg.norm(X) # Same function for L2 Norm
```

```python
import torch 

X_pt = torch.tensor([[1,2],[3,4.]]) # torch.norm() requires float
```

```python
torch.norm(X_pt)
```
