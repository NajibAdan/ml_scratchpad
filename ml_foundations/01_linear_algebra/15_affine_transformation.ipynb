# Affine Transformation

They change the geometry,angles or distances of the matrices but preserves parallelism between them

You can scale, flip, shear or rotating the matrix.

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import plot_vectors
v = np.array([3,1])
v
```

```python
plot_vectors([v],['lightblue'])
plt.xlim(-1,5)
plt.ylim(-1,5);
```

```python
I = np.array([[1,0],[0,1]])
I
```

```python
Iv = np.dot(I,v)
Iv
```

```python
v == Iv
```

```python
plot_vectors([Iv],['blue'])
plt.xlim(-1,5)
plt.ylim(-1,5)
```

```python
E = np.array([[1,0],[0,-1]])
E
```

```python
Ev = np.dot(E,v)
Ev
```

```python
plot_vectors([v,Ev],['lightblue','blue'])
plt.xlim(-1,5)
plt.ylim(-1,5);
```

```python
F = np.array([[-1,0],[0,1]])
F
```

```python
Fv = np.dot(F,v)
Fv
```

```python
plot_vectors([v,Fv],['lightblue','blue'])
plt.xlim(-4,4)
plt.ylim(-1,5);
```
