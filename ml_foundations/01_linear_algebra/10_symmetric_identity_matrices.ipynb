# Symmetric Matrices

Speacial matrix case with the following properties:
- Square
- Transpose of X == X


```python
import numpy as np

X_sym = np.array([[0,1,2],[1,7,8],[2,8,9]])
X_sym
```

```python
X_sym.T == X_sym
```

# Identity Matrix
Symmetrix where
- Every element along the main diag is 1
- All the other elements are 0
- Notation: **I<sub>n</sub>** where *n* = height (or width)
- n-length vector remains unchanged if multiplied by **I<sub>n</sub>**



```python
import torch 
I = torch.tensor([[1,0,0],[0,1,0],[0,0,1]])
I
```

```python
x_pt = torch.tensor([25,2,5])
x_pt
```

```python
torch.matmul(I,x_pt)
```
